```python
%matplotlib inline
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATASET 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
# Create a bar chart based upon the above data
plt.bar(gyms, members, color="b", align="center")
plt.show()
```

```python
# DATASET 2
x_lim = 2 * np.pi
x_axis = np.arange(0, x_lim, 0.1)
sin = np.sin(x_axis)
```

```python
#Creating a line graph
plt.plot(x_axis, sin)
plt.show()
```

```python
# DATASET 3
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
x_axis = np.arange(0, len(gyms))
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue"]
explode = (0, 0.05, 0, 0)
```

```python
plt.pie(members, explode=explode,labels=gyms, colors=colors,autopct="%1.1f%%", startangle=140)
plt.axis("equal")
```

```python
# DATASET 4
x_axis = np.arange(0, 10, 0.1)
times = []
for x in x_axis:
    times.append(x * x + np.random.randint(0, np.ceil(max(x_axis))))
```

```python
plt.scatter(x_axis,times, marker="o", s=times, alpha=0.75, edgecolors="black")
plt.title("Random Dots")
```
